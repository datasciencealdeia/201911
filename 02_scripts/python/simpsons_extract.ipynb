```python
#Processamento para todos arquivos, gerar vetor de caracteristica por imagem teste e train

import numpy as np
import os
from PIL import Image #Para trabalhar com imagem, simples, roots

#Funcao para retornar a imagem como array de pixels
def get_image_pixels(image, width, height):
    """Get a numpy array of an image so that one can access values[x][y]."""
    pixel_values = list(image.getdata())
    if image.mode == 'RGB':
        channels = 3
    elif image.mode == 'L':
        channels = 1
    else:
        print("Unknown mode: %s" % image.mode)
        return None
    pixel_values = np.array(pixel_values).reshape((width, height, channels))
    return pixel_values

def ehLaranja (r, g, b):
    if 232 <= r <= 247 and 91 <= g <= 99 and 11 <= b <= 22:
        return True
    else:
        return False

def ehAzulEscuro (r, g, b):
    if 0 <= r <= 16 and 8 <= g <= 17 and 132 <= b <= 177:
        return True
    else:
        return False

def ehBranco (r, g, b):
    if 250 <= r <= 255 and 250 <= g <= 255 and 250 <= b <= 255:
        return True
    else:
        return False

def ehMarrom (r, g, b):
    if 166 <= r <= 206 and 153 <= g <= 173 and 91 <= b <= 155:
        return True
    else:
        return False

def ehAzulJeans (r, g, b):
    if 0 <= r <= 50 and 64 <= g <= 107 and 173 <= b <= 177:
        return True
    else:
        return False

#Funcao principal do nosso codigo
def main():
    dir_desafio = '/home/ds/git/201909/03_dados/desafio pessoal/'
    dir_name_treino = dir_desafio + 'train/'
    dir_name_teste = dir_desafio + 'test/'
    vetorTreino = []
    vetorTeste = []
    iContFile = 0
    
    imagesTreino = os.listdir(dir_name_treino) 
    imagesTeste = os.listdir(dir_name_teste) 
    
    for imgTreino in imagesTreino:
        iLaranja, iBranco, iAzulEscuro, iAzulJeans, iMarrom, iOutros = 0, 0, 0, 0, 0, 0
        iContFile += 1

        print (iContFile, end="")
        print ("\r", end="")
        
        im = Image.open(dir_name_treino+imgTreino, 'r')
        largura, altura = im.size

        imagem = get_image_pixels(im, largura, altura)

        for i in range(largura):
            for y in range (altura):
                pixel_r = imagem[i][y][0]
                pixel_g = imagem[i][y][1]
                pixel_b = imagem[i][y][2]

                """
                    Faixa de Cores RGB
                    Bart
                        Laranja R(232,247) ,G(91,99), B(11,22)
                        Azul Escuro R(0, 16), G(8, 17), B(132,177)
                    Homer
                        Branco R(250,255), G(250,255), B(250,255)
                        Marron R(166, 206), G(153, 173), B(91, 155)
                        Azul Jeans R(0, 50), G(64, 107), B(173,177)
                """

                #Contador cores

                if ehLaranja(pixel_r, pixel_g, pixel_b):
                    iLaranja += 1
                elif ehAzulEscuro(pixel_r, pixel_g, pixel_b):
                    iAzulEscuro += 1
                elif ehAzulJeans(pixel_r, pixel_g, pixel_b):
                    iAzulJeans += 1
                elif ehMarrom(pixel_r, pixel_g, pixel_b):
                    iMarrom += 1
                elif ehBranco(pixel_r, pixel_g, pixel_b):
                    iBranco += 1
                else:
                    iOutros += 1
        
        fPercLaranja = (iLaranja / (largura * altura))
        fPercBranco = (iBranco / (largura * altura))
        fPercAzulEscuro = (iAzulEscuro / (largura * altura))
        fPercAzulJeans = (iAzulJeans / (largura * altura))
        fPercMarrom = (iMarrom / (largura * altura))
        fPercOutros = (iOutros / (largura * altura))
        strRotulo = "bart" if imgTreino.startswith("bart") else "homer"

        vetorTreino.append([fPercLaranja,fPercBranco,fPercAzulEscuro,fPercAzulJeans,fPercMarrom,fPercOutros,strRotulo])
           
        """"
        print ("Laranja %f" )
        print ("Branco %f" %(iBranco / (largura * altura)))
        print ("Azul Escuro %f" %(iAzulEscuro / (largura * altura)))
        print ("Azul Jeans %f" %(iAzulJeans / (largura * altura)))
        print ("Marrom %f" %(iMarrom / (largura * altura)))
        print ("Outros %f" %(iOutros / (largura * altura)))
        """
    np.savetxt(dir_desafio+'treino.csv', vetorTreino, delimiter=";", fmt="%s")

    iContFile = 0
    print ("   ", end="")
    
    for imgTeste in imagesTeste:
        iLaranja, iBranco, iAzulEscuro, iAzulJeans, iMarrom, iOutros = 0, 0, 0, 0, 0, 0
        iContFile += 1

        print (iContFile, end="")
        print ("\r", end="")
        
        im = Image.open(dir_name_teste+imgTeste, 'r')
        largura, altura = im.size

        imagem = get_image_pixels(im, largura, altura)

        for i in range(largura):
            for y in range (altura):
                pixel_r = imagem[i][y][0]
                pixel_g = imagem[i][y][1]
                pixel_b = imagem[i][y][2]

                """
                    Faixa de Cores RGB
                    Bart
                        Laranja R(232,247) ,G(91,99), B(11,22)
                        Azul Escuro R(0, 16), G(8, 17), B(132,177)
                    Homer
                        Branco R(250,255), G(250,255), B(250,255)
                        Marron R(166, 206), G(153, 173), B(91, 155)
                        Azul Jeans R(0, 50), G(64, 107), B(173,177)
                """

                #Contador cores

                if ehLaranja(pixel_r, pixel_g, pixel_b):
                    iLaranja += 1
                elif ehAzulEscuro(pixel_r, pixel_g, pixel_b):
                    iAzulEscuro += 1
                elif ehAzulJeans(pixel_r, pixel_g, pixel_b):
                    iAzulJeans += 1
                elif ehMarrom(pixel_r, pixel_g, pixel_b):
                    iMarrom += 1
                elif ehBranco(pixel_r, pixel_g, pixel_b):
                    iBranco += 1
                else:
                    iOutros += 1
        
        fPercLaranja = (iLaranja / (largura * altura))
        fPercBranco = (iBranco / (largura * altura))
        fPercAzulEscuro = (iAzulEscuro / (largura * altura))
        fPercAzulJeans = (iAzulJeans / (largura * altura))
        fPercMarrom = (iMarrom / (largura * altura))
        fPercOutros = (iOutros / (largura * altura))
        strRotulo = "bart" if imgTeste.startswith("bart") else "homer"

        vetorTeste.append([fPercLaranja,fPercBranco,fPercAzulEscuro,fPercAzulJeans,fPercMarrom,fPercOutros,strRotulo])
           
        """"
        print ("Laranja %f" )
        print ("Branco %f" %(iBranco / (largura * altura)))
        print ("Azul Escuro %f" %(iAzulEscuro / (largura * altura)))
        print ("Azul Jeans %f" %(iAzulJeans / (largura * altura)))
        print ("Marrom %f" %(iMarrom / (largura * altura)))
        print ("Outros %f" %(iOutros / (largura * altura)))
        """
    np.savetxt(dir_desafio+'teste.csv', vetorTeste, delimiter=";", fmt="%s")

   
#Best Practice - para definir onde comeca a execucao do nosso codigo
if __name__ == "__main__":
    main()
```
